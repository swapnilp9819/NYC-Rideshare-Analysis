```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Loading the datasets from generated CSV files
trips_df = pd.read_csv('task02output/que1/part-00000-575f6965-c33c-442c-946e-eec3163ce0c2-c000.csv')
profits_df = pd.read_csv('task02output/que2/part-00000-c62201b5-3b22-4fa1-ab63-ad72f6c69ad8-c000.csv')
earnings_df = pd.read_csv('task02output/que3/part-00000-cbf9e611-6824-499a-8a27-5851cfce48ee-c000.csv')

# Sort the data by 'business-month' in ascending order for all DataFrames
trips_df.sort_values(by='business_month', inplace=True)
profits_df.sort_values(by='business_month', inplace=True)
earnings_df.sort_values(by='business_month', inplace=True)

# Trips per Business Month Histogram
plt.figure(figsize=(14, 7))
plt.bar(trips_df['business_month'], trips_df['trip_count'], color='skyblue')
plt.yscale('log')  # Set the y-axis to a log scale
plt.xlabel('Business-Month')
plt.ylabel('Number of Trips (log scale)')
plt.title('Number of Trips for Each Business in Each Month (log scale)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Profits per Business Month Histogram
plt.figure(figsize=(14, 7))
plt.bar(profits_df['business_month'], profits_df['total_profit'], color='lightgreen')
plt.yscale('log')  # Set the y-axis to a log scale
plt.xlabel('Business-Month')
plt.ylabel('Platform Profits (log scale)')
plt.title('Platform Profits for Each Business in Each Month (log scale)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Earnings per Business Month Histogram
plt.figure(figsize=(14, 7))
plt.bar(earnings_df['business_month'], earnings_df['total_earnings'], color='salmon')
plt.yscale('log')  # Set the y-axis to a log scale
plt.xlabel('Business-Month')
plt.ylabel('Driver Earnings (log scale)')
plt.title('Driver Earnings for Each Business in Each Month (log scale)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
